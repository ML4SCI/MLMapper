```python
import sys
sys.path.extend([r'/Users/peplopn1/python/pds4_tools-1.3/'])
import pds4_tools

import glob
import numpy as np
import matplotlib.pyplot as plt
```

## Identify the PDS files

```python
work_dir = '/Users/peplopn1/project/Lunar Prospector/'
data_dir = 'PIXON_HEGR_Mg_Map/GRS_Calibrated_Time_Series_Spectra/Data/'

files    = sorted(glob.glob(work_dir+data_dir+'199*_***_grs.xml'))  

index    = np.arange(0, len(files), 1)
```

## Create the energy axis for the spectra

```python
gain     = 17.8 #keV/channel
energy   = np.arange(0, 512, 1)*gain

print('Min, Max energy = ', round(min(energy)), ', ', round(max(energy)), ' keV')
```

## Read/parse the PDS data from each daily file, and concatenate into a single time series

```python
for p in index:
    structures = pds4_tools.read(files[np.int(p)])
    #structures.info()

    table = structures[0]
    ##table.info()

    Acc_Spec_temp = table.fields[0] 
    Rej_Spec_temp = table.fields[1] 
    Deadtime_temp = table.fields[2] 
    Overload_temp = table.fields[3] 
    Alt_temp      = table.fields[6] 
    Lat_temp      = table.fields[7] 
    Lon_temp      = table.fields[8] 
    
    if p == 0:
        Acc_Spec = Acc_Spec_temp 
        Rej_Spec = Rej_Spec_temp 
        Deadtime = Deadtime_temp 
        Overload = Overload_temp 
        Alt      = Alt_temp      
        Lat      = Lat_temp      
        Lon      = Lon_temp      
    if p > 0:
        Acc_Spec = np.concatenate((Acc_Spec, Acc_Spec_temp)) 
        Rej_Spec = np.concatenate((Rej_Spec, Rej_Spec_temp)) 
        Deadtime = np.concatenate((Deadtime, Deadtime_temp)) 
        Overload = np.concatenate((Overload, Overload_temp)) 
        Alt      = np.concatenate((Alt,      Alt_temp))      
        Lat      = np.concatenate((Lat,      Lat_temp))      
        Lon      = np.concatenate((Lon,      Lon_temp))  
```

## Make the deadtime-corrected High-Energy Gamma Ray (HEGR) time series

```python
fig  = plt.figure(figsize=(12,8), dpi=60)

plt.xlabel('Energy', size=35)
plt.ylabel('Counts/32 s', size=35)
plt.tick_params(axis='both', which='major', labelsize=30)
plt.yscale('log')

y = np.sum(Acc_Spec,0)/len(Alt)

lines  = np.array([511, 846, 1460, 2615, 4438, 6128, 7635])
index2 = np.arange(0, len(lines), 1) 
for q in index2:
    plt.plot([lines[int(q)], lines[int(q)]], [1e-10, 1e10], linestyle='dashed', color='blue')

plt.plot([8000., 8000.], [1e-10, 1e10], linestyle='dashed', color='red')
plt.plot([8900., 8900.], [1e-10, 1e10], linestyle='dashed', color='red')

plt.plot(energy, y, figure=fig, color='black', drawstyle='steps')
plt.axis([0., 9150, 0.4, 1.2*max(y)])   
plt.show()

gate = (energy > 8000)&(energy < 8900)
HEGR_Cnts = np.sum(Acc_Spec[::,gate], axis=1)/(32./(1-Deadtime))
HEGR_Rate = HEGR_Cnts*60
HEGR_Err  = np.sqrt(HEGR_Cnts)*60

gate = (energy > 100)&(energy < 8800)
All_Cnts  = np.sum(Acc_Spec[::,gate], axis=1)/(32./(1-Deadtime))
All_Rate  = All_Cnts*60
All_Err   = np.sqrt(All_Cnts)*60
```

## Plot the HEGR and altitude time series as a check of the data

```python
x = np.arange(0, len(Alt), 1)

fig  = plt.figure(figsize=(10,6), dpi=60)
plt.xlabel('Spectrum Index', size=30)
plt.ylabel('Altitude (km)', size=30)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')
plt.plot(x, Alt, figure=fig, color='black')
#plt.axis([2000., 3000, 1./Period[0], 1.6*max(Spectrum)])   
plt.show()

fig  = plt.figure(figsize=(10,6), dpi=60)
plt.xlabel('Spectrum Index', size=30)
plt.ylabel('High-Energy Gamma Rays (cpm)', size=30)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')
plt.plot(x, HEGR_Rate, figure=fig, color='black')
#plt.axis([2000., 3000, 1./Period[0], 1.6*max(Spectrum)])   
plt.show()

fig  = plt.figure(figsize=(10,6), dpi=60)
plt.xlabel('Spectrum Index', size=30)
plt.ylabel('All Gamma Rays (cpm)', size=30)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')
plt.plot(x, All_Rate, figure=fig, color='black')
#plt.axis([2000., 3000, 1./Period[0], 1.6*max(Spectrum)])   
plt.show()

fig  = plt.figure(figsize=(10,6), dpi=60)
plt.xlabel('Spectrum Index', size=30)
plt.ylabel('All Gamma Rays (cpm)', size=30)
plt.tick_params(axis='both', which='major', labelsize=25)
#plt.yscale('log')
plt.plot(x, Lat, figure=fig, color='black')
plt.axis([32500, 40000, -90, 90])   
plt.show()
```
